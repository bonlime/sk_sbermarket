```python
GLOBAL_COUNTER = 0
def apply_func(x):
    # x is sub df with columns parent_category_id & count
    # how many items to take from each category
    x["count"] = (x["count"] / x["count"].sum() * 70).astype(int)
    # fill with most `count` from each category 
    res = []
    for row in x.itertuples():
        res.extend(grouped_by_pcat_prod.loc[row.parent_category_id].index[:row.count])
        
    global GLOBAL_COUNTER
    GLOBAL_COUNTER += 1
#     if GLOBAL_COUNTER < 0:        
#         raise ValueError
#     print(res)
    if len(res) < 50:
        print(f"AHTUNG {GLOBAL_COUNTER:06}")
    return " ".join([str(i) for i in res[:50]])
    
new_subm2 = grouped_by_pcat.groupby('user_id')[["parent_category_id", "count"]].apply(apply_func).reset_index()
```
